```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset
dataset_path = r'C:\Users\SODIKAP\Pictures\dataset\water_potability.csv'
df = pd.read_csv(dataset_path)

# Menampilkan informasi dataset
print("Informasi dataset:")
print(df.info())

# Menampilkan lima baris pertama dari dataset
print("\nLima baris pertama dari dataset:")
print(df.head())

# Statistik deskriptif
print("\nStatistik deskriptif:")
print(df.describe())

# Menghitung jumlah nilai yang hilang pada setiap kolom
print("\nJumlah nilai yang hilang pada setiap kolom:")
print(df.isnull().sum())

# Menampilkan korelasi antar-variabel
print("\nKorelasi antar-variabel:")
print(df.corr())

# Menampilkan distribusi kelas pada kolom "Potability"
print("\nDistribusi kelas pada kolom 'Potability':")
print(df['Potability'].value_counts())

# Menampilkan grafik distribusi variabel "ph"
df['ph'].hist()
plt.xlabel('pH')
plt.ylabel('Count')
plt.title('Distribusi pH')
plt.show()

# Menampilkan scatter plot antara variabel "ph" dan "Hardness"
plt.scatter(df['ph'], df['Hardness'])
plt.xlabel('pH')
plt.ylabel('Hardness')
plt.title('Scatter Plot pH vs Hardness')
plt.show()

# Menampilkan box plot variabel "Potability" terhadap variabel "Hardness"
sns.boxplot(x='Potability', y='Hardness', data=df)
plt.xlabel('Potability')
plt.ylabel('Hardness')
plt.title('Box Plot Potability vs Hardness')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca dataset
dataset_path = r'C:\Users\SODIKAP\Pictures\dataset\water_potability.csv'
df = pd.read_csv(dataset_path)

# Membagi dataset berdasarkan potabilitas
non_potable = df.query("Potability == 0")
potable = df.query("Potability == 1")

# Membuat plot KDE untuk setiap fitur dan menyimpan gambar
plt.figure(figsize=(15, 15))
for col in df.columns[:9]:
    plt.subplot(3, 3, df.columns.get_loc(col) + 1)
    plt.title(col)
    sns.kdeplot(x=non_potable[col], label="Non Potable")
    sns.kdeplot(x=potable[col], label="Potable")
    plt.legend()
    plt.savefig(f"KDE_{col}.png")  # Menyimpan gambar

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca dataset
dataset_path = r'C:\Users\SODIKAP\Pictures\dataset\water_potability.csv'
df = pd.read_csv(dataset_path)

# Daftar variabel numerik
numvars = ["ph", "Hardness", "Solids", "Chloramines", "Sulfate", "Conductivity", "Organic_carbon", "Trihalomethanes", "Turbidity"]

# Membuat histogram KDE dengan sumbu Y sebagai density dan warna mencolok
color_palette = ["#FF5A5F", "#00A699"]  # Warna untuk kategori "Non Potable" dan "Potable"
for var in numvars:
    plt.figure(figsize=(10, 6))
    sns.kdeplot(x=df.loc[df['Potability'] == 0, var], color=color_palette[0], label="Non Potable")
    sns.kdeplot(x=df.loc[df['Potability'] == 1, var], color=color_palette[1], label="Potable")
    plt.title(f"Faktor-faktor yang Mempengaruhi Kualitas Air: {var}")
    plt.xlabel(var)
    plt.ylabel("Density")
    plt.legend(title="Potability", labels=["Non Potable", "Potable"])
    plt.savefig(f"C:/Users/SODIKAP/Pictures/KDE_{var}.png")  # Menyimpan gambar
    plt.show()
```
